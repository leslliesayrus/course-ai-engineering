```python
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.2, #0.7
    api_key = api_key

    )
```

```python
texto = """Joao e um cara de 20 anos,
ele e muito bom em Python e em IA"""
```

```python
from pydantic import BaseModel, Field
class BotResponse(BaseModel):
    nome: str = Field(description="nome da pessoa")
    idade: int = Field(description="idade da pessoa")
    habilidades: list[str] = Field(description="habilidades")
    brasileiro: bool = Field(default = False, description="É um brasileiro?")
```

```python
structured_llm = llm.with_structured_output(BotResponse)
```

```python
result = structured_llm.invoke(texto)
```

```python
result
```

```python
paises = ["Brasil", "EUA", "Inglaterra", "Japao", "Argentina"]
```

```python
class PaisNascimento(BaseModel):
    pais: str = Field(
        ...,
        enum=paises,
        description="escolha um dos paises",
    )
```

```python
structured_llm = llm.with_structured_output(PaisNascimento)
```

```python
structured_llm.invoke("Onde nasceu o Thomas?")
```

# Parâmetros dos LLM's

---

## Parâmetros principais

- **model**: Nome do modelo usado.  
- **temperature**: Aleatoriedade da resposta (0 a 1).  
- **max_tokens**: Limite total de tokens (entrada + saída).  
- **max_completion_tokens**: Limite de tokens da resposta gerada.  

---

## Raciocínio (reasoning)

- **reasoning_effort**: Nível de esforço de raciocínio (`low` / `medium` / `high`).  
- **reasoning_format**: Formato de retorno do raciocínio (`parsed` / `raw` / `hidden`).  

---

## Rede e controle

- **timeout**: Tempo máximo de espera da resposta (s).  
- **max_retries**: Número máximo de tentativas em caso de falha.  
- **api_key**: Chave de autenticação para o serviço Groq.
